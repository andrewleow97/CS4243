### Template Matching Demo
(https://https://docs.opencv.org/2.4/doc/tutorials/imgproc/histograms/template_matching/template_matching.html#which-are-the-matching-methods-available-in-opencv)

For reference to what different methods do to match the correct image refer to https://docs.opencv.org/2.4/doc/tutorials/imgproc/histograms/template_matching/template_matching.html#which-are-the-matching-methods-available-in-opencv

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show_method_result(method, row_number_to_show, p_img_file_name, 
                       p_template_file_name, method_name):
    p_img = cv2.imread(p_img_file_name, 0)
    p_template = cv2.imread(p_template_file_name,0)
    w, h = p_template.shape[::-1]

    res = cv2.matchTemplate(p_img, p_template, method)
    print("Shape of correlation result ", res.shape)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)
    cv2.rectangle(p_img, top_left, bottom_right, 255, 2)

    plt.subplot(7, 2, row_number_to_show * 2 - 1),plt.imshow(res, cmap = 'gray')
    plt.title('Matching Result ' + str(method_name)), plt.xticks([]), plt.yticks([])
    plt.subplot(7, 2, row_number_to_show * 2),plt.imshow(p_img,cmap = 'gray')
    plt.title('Detected Point ' + str(method_name)), plt.xticks([]), plt.yticks([])

plt.figure(figsize=(8, 20))
img = cv2.imread('MainImage.png',0)
template = cv2.imread('Template.png',0)
print("Image Shape ", img.shape)
print("Template shape", template.shape)
plt.subplot(3, 2, 1), plt.imshow(img, cmap='gray'), plt.axis("off")
plt.subplot(3, 2, 2), plt.imshow(template, cmap='gray'), plt.axis("off")
w, h = template.shape[::-1]

show_method_result(cv2.TM_CCORR, 4, 'MainImage.png', 'Template.png', 'TM_CCORR')
show_method_result(cv2.TM_CCORR_NORMED, 5, 'MainImage.png', 'Template.png', 'TM_CCORR_NORMED')
```
