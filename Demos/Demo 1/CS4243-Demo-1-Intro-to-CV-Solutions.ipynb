```python
def hist_stretch(image):
  
    maxpix = np.max(image)
    minpix = np.min(image)

    image = 255.0 * (image - minpix) / (maxpix - minpix)
    return image.astype('uint8')

def hist_equalize(image):
    # np.histogram(image.flatten(), 256, [0,255])
    hist, bins = np.histogram(image.flatten(), 256, [0,256]) # 
    cdf = hist.cumsum()
    cdf_normalized = cdf / cdf.max()

    plt.plot(cdf_normalized * hist.max(), color = 'b')
    plt.hist(image.flatten(), 256, [0,256], color = 'r')
    for color in range(256):
        image = np.where(image==color, cdf_normalized[color], image)
    return (image * 255.0).astype('uint8')
```

```python
def load_convert_resize(image_name, resize=False):
    img = cv2.imread(image_name)
    img_c = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    if resize is True:
        img_c = cv2.resize(img_c, (512, 512))
    return img_c

def plot_images(image_array, title, fig_size1=16, fg_size2=16):
    plt.figure(figsize=(fig_size1, fg_size2))

    for i in range(len(image_array)):
        plt.subplot(len(image_array), 1, i + 1)
        plt.imshow(image_array[i], cmap = plt.get_cmap(name="gray"))
        plt.title(title[i])

#Image Normalization (Whitening)
def whitening(image):
    image = image.astype('float64')
    mean = np.mean(image)
    std = np.std(image)
    image = ((image - mean) / std)
    #image = hist_stretch(image)
    return image
```

Start Image = [4, 7, 4, 10, 15, 15]
calculating Histogram: [2, 1, 1, 2], [4, 7, 10, 15] --- np.histogram

Calculating CDF: [2, 3, 4, 6], [4, 7, 10, 15] --- .cumsum()

Normalizing CDF Values: [2/6, 3/6, 4/6, 6/6], [4, 7, 10, 15] -- /max()

New Values of Image pixels: [0.3, 0.5, 0.7, 1], [4, 7, 10, 15]

Substituting Values in the New image: [0.3, 0.5, 0.3, 0.7, 1, 1] np.where()

Converting Image to Uint: [76, 127, 76, 178, 255, 255]

```python
def exp_map(img):
  
    img = img.astype('float64')
    maxpix = np.max(img)
#   new_image = (255.0/np.power(maxpix, 10)) * np.power(img, 10)
    new_image = (255.0/(np.power(1.03, maxpix) - 1)) * (np.power(1.03, img) - 1)

    return new_image.astype('uint8')
  
def log_map(img):
  
    img = img.astype('float64')
    maxpix = np.max(img)
    new_image = (255.0/np.log(1 + maxpix)) * (np.log(1 + img))

    return new_image.astype('uint8')
```

```python
image1 = cv2.imread('car_log.jpg')
image2 = cv2.imread('game_exp.jpg')

image1_log = log_map(image1)
image2_exp = exp_map(image2)

plt.scatter(image1.flatten(), image1_log.flatten())
plt.scatter(image2.flatten(), image2_exp.flatten())

plt.figure(figsize=(20, 10))
plt.subplot(1,4,1)
plt.imshow(image1)
plt.title('Original image 1')

plt.subplot(1,4,2)
plt.imshow(image1_log)
plt.title('Log mapped image 1')

plt.subplot(1,4,3)
plt.imshow(image2)
plt.title('Original image 2')

plt.subplot(1,4,4)
plt.imshow(image2_exp)
plt.title('Exp mapped image 2')
```
